```python
import pandas as pd
import numpy as np
 
data = {
    "Customer_Name": ["Ram", "Kishore", "Deva", "Moni"],
    "Order_ID": [101,102,103,104],
    "Product": ["Laptop","Mobile","Mouse","Keyboard"],
    "Quantity": [1,4,5,5],
    "Price": [50000,60000,8000,10000]
}
 
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

avg_sales = df.groupby("Product")["Price"].mean()

avg_sales.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Average Sales")
plt.xlabel("Product")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```
